### Exploratory Analysis of  Dataset

```python
#Import pandas, seaborn and matplotlib libraries
#Create a dataframe
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df = pd.read_csv('https://raw.githubusercontent.com/frankData612/data_612/master/baseballdatabank-master/core/HomeGames.csv')
```

```python
#Scatter plot of games by attendance with color by league
scatter, ax = plt.subplots()
ax = sns.scatterplot(x='games',y='attendance', hue='league.key', data=df)
ax.set_title('Scatterplot of Games by Attendance by League')
ax.set_xlabel('Games')
ax.set_ylabel('Attendance')
plt.show()
```

```python
#Kernel Density chart of attendance by year
kde, ax = plt.subplots()
ax = sns.kdeplot(x='year.key', y='attendance', data=df, shade=True)
ax.set_title('Kernel Density Plot of Attendance by Year')
ax.set_xlabel('Year')
ax.set_ylabel('Attendance')
plt.show()
```

```python
#Distribution of games by league in a boxplot
box, ax = plt.subplots()
ax = sns.boxplot(x='league.key', y='games', data=df)
ax.set_title('Boxplot of Games by League')
ax.set_xlabel('League')
ax.set_ylabel('Games')
plt.show()
```
