### Create a function to compare two integers

```python
def total(num1,num2):
    #create a function that compares two integers
    #create two variables, total and list
    #total variable to calculate total of the numbers between integers
    #list variable to create a list to hold both integers
    total = 0
    list = [num1, num2]
    for num in list:
        if list[0] > list[1]:
        #if num1 is greater than num2
        #print a statement stating the total between is 0
            print('The total of the numbers in between is 0')
            break
        elif list[0] == list[1]:
        #if num1 is equal to num2
        #print a statement stating the total between is the value of the first integer
            print('The total of the numbers in between is', list[0])
            break
        else:
        #if num1 is less than num2
        #calculate the total of the numbers between the two variables
        #print a statement stating the total between is the value of the total variable
            for num in range(list[0],list[1]+1):
                total += num
            print('The total of the numbers in between is', total)
            break

#prompt the user to input two integers
#call the total function                  
num1 = int(input('Enter the first number:'))
num2 = int(input('Enter the second number:'))
total(num1,num2)
```

### Create a function to return statistics on a dataframe

```python
#Create a function that returns the mean, sum, mode, median, and range (separately)
#Run the function into your chosen data set using the .apply() method.
from scipy import stats
import numpy as np

df = pd.read_csv('https://raw.githubusercontent.com/frankData612/data_612/master/State_Drug_Utilization_Data_2010/State_Drug_Utilization_Data_2010.csv')

def descrip_stats(x):
    """Return the mean, sum, mode, median and range"""
    mean = np.mean(x)
    sum1 = np.sum(x)
    mode = stats.mode(x)
    median = np.median(x)
    range1 = (np.amax(x) - np.amin(x))
    return (mean,sum1,mode,median,range1)

results = (df[['Number of Prescriptions']].apply(descrip_stats))
results1 = results.iloc[2,0]
print('The mean is {}'.format(results.iloc[0, 0]))
print('The sum is {}'.format(results.iloc[1,0]))
print('The mode is {}'.format(results1[0]))
print('The median is {}'.format(results.iloc[3,0]))
print('The range is {}'.format(results.iloc[4,0]))
```

### Create a function to provide summary statistics on a dataframe

```python
homegames_df = pd.read_csv('https://raw.githubusercontent.com/frankData612/data_612/master/baseballdatabank-master/core/HomeGames.csv')
parks_df = pd.read_csv('https://raw.githubusercontent.com/frankData612/data_612/master/baseballdatabank-master/core/Parks.csv')
combined_df = homegames_df.merge(parks_df, left_on='park.key', right_on='park.key')
combined_df = combined_df.fillna(method='ffill')
combined_df = combined_df.fillna(method='bfill')

combined_df.rename({'year.key':'year', 'league.key':'league', 'team.key':'team', 'park.key':'park_key', 
                    'span.first':'span_first','span.last':'span_last', 'games':'games', 'openings':'openings',
                    'attendance':'attendance', 'park.name':'park_name', 'park.alias':'park_alias', 'city':'city', 
                    'state':'state', 'country':'country'}, axis='columns', errors='raise', inplace=True)

def summarize_data(x):
    return (pd.Series({
            'avg_games': x.games.mean(),
            'min_games': x.games.min(),
            'max_games': x.games.max(),
            'std_dev_games': x.games.std(),
            '1st_quartile': x.games.quantile(q=0.25),
            '2nd_quartile': x.games.quantile(q=0.50),
            '3rd_quartile': x.games.quantile(q=0.75),
            'sample_count': len(x),
            'proportion': x.games.count()/combined_df.games.count()
        }))

combined_df.groupby('league').apply(summarize_data)
```
