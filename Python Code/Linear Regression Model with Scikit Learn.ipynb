### Linear Regression with Scikit Learn

#### Load the dataset

```python
#import the pandas library
import pandas as pd

#create a variable df and import the mpg.csv file using the read_csv function
df = pd.read_csv('https://raw.githubusercontent.com/franklin-univ-data-science/data/master/mpg.csv')
```

#### Remove all NaN values from the dataset

```python
#Remove all NaN values from the dataset
#call the dropna function on the dataframe and assign to the df variable
df = df.dropna()

#confirm the NaN values were dropped fromt he dataset
#print the data type and count of each variable in the dataset by calling the info function on the df variable
df.info()
```

####  Plot the scatter plot matrix and correlation matrix, pick two predictor variables for linear regression

##### Pairplot Visualization

```python
#import the pyplot submodule from the Matplotlib library
#import the seaborn library
import matplotlib.pyplot as plt
import seaborn as sns

#call the pairplot function from the seaborn library
##in the function provide the df variable and assign the height parameter with the value of 2.5
#call the tight_layout function from the matplotlib library
#call the show function from the matplotlib library
sns.pairplot(df, height=2.5)
plt.tight_layout()
plt.show()
```

##### Confusion Matrix

```python
import numpy as np

#create the variable cm, call the correlation coefficient function from the numpy library
##in the function pass in the df variable values and transpose by calling the T function
#call the set function on the seaborn library and assign the font_scale parameter with the value of 1.5
#create the hm variable and assign by calling the heatmap function from the seaborn library
##pass in the cm variable
##assign the cbar parameter as True
##assign the annot parameter as True
##assign the fmt parameter as '.2f' - two decimal points
##assign the annot_kws parameter with key='size', value=15
##assign the yticklabels parameter with the df columns names
##assign the xticklabels parameter with the df columns names
##call the show function from the matplotlib library
cm = np.corrcoef(df.values.T)
sns.set(font_scale=1.5)
hm = sns.heatmap(cm, cbar=True, annot=True, square=True, fmt='.2f', annot_kws={'size':15}, 
                 yticklabels=df.columns, xticklabels=df.columns)
plt.show()
```

#### Create a dataframe with the two selected variables

```python
#from the scikit learn library model selection submodule import the train_test_split module
#from the scikit learn library linear model submodule import the Linear Regression module
#from the scikit learn library metrics submodule import the mean_squared_error and r2_score modules
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

#create the feature matrix X and assign with the df 'cylinders' and 'displacement' column values
#create the target variable y and assign with the df 'mpg' column values
X = df[['cylinders', 'displacement']].values
y = df[['mpg']].values
```

#### Split the dataset into a training and test dataset

```python
#split the dataset using train_test_split
#assign the X_train, X_test, y_train, y_test variables using train_test_split
##pass in the feature matrix - X, and the target vector - y
##assign the test_size parameter with 0.2 or 20% of the dataset
#assign the random_state parameter with a value of 1 so our results are the same every time
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
```

#### Create a Linear Regression model from scikit learn

```python
#create the lr variable and assign with the LinearRegression function
#call the fit function on the lr variable and pass in the X_train and y_train datasets
#create the y_train_pred variable
##assign the values by calling the predict function on the lr variable and pass in the X_train dataset
#create the y_test_pred variable
##assign the values by calling the predict function on the lr variable and pass in the X_test dataset
lr = LinearRegression()
lr.fit(X_train, y_train)
y_train_pred = lr.predict(X_train)
y_test_pred = lr.predict(X_test)
```

#### Print the Mean Squared Error for the Linear Regression Model

```python
#print the Mean Squared Error of the training set and test set
#provide the string to print and format the results to 3 decimal places
#call the mean squared error function and pass in the y_train and y_train_pred datasets
#call the mean squared error function and pass in the y_test and y_test_pred datasets
print('Mean Squared Error Train: {:.3f} \nMean Squared Error Test: {:.3f}'.format(mean_squared_error(y_train, y_train_pred),
                                                  mean_squared_error(y_test, y_test_pred)))
```

#### Print the R$^2$ value for the Linear Regression Model

```python
#print the R-Squared of the training set and test set
#provide the string to print and format the results to 3 decimal places
#call the r2_score function and pass in the y_train and y_train_pred datasets
#call the r2_score function and pass in the y_test and y_test_pred datasets
print('R^2 Train: {:.3f} \nR^2 Test: {:.3f}'.format(r2_score(y_train, y_train_pred), 
                                                   r2_score(y_test, y_test_pred)))
```

#### Plot the Residuals for the Training and Test Datasets

```python
#plot the residuals of the train set
#call the scatter function from the Matplotlib library
##for the x-axis provide the y_train_pred dataset
##for y-axis provide the difference between the y_train_pred and y_train datasets
##for the c parameter assign the value 'steelblue'
##for the marker parameter assign the value 'o'
##for the edgecolor parameter assign the value 'white'
##for the label parameter assign the value 'Training data'
plt.scatter(y_train_pred, y_train_pred - y_train, c='steelblue', marker='o', edgecolor='white', label='Training data')

#plot the residuals of the test set
#call the scatter function from the Matplotlib library
##for the x-axis provide the y_test_pred dataset
##for y-axis provide the difference between the y_test_pred and y_test datasets
##for the c parameter assign the value 'limegreen'
##for the marker parameter assign the value 's'
##for the edgecolor parameter assign the value 'white'
##for the label parameter assign the value 'Test data'
plt.scatter(y_test_pred, y_test_pred - y_test, c='limegreen', marker='s', edgecolor='white', label='Test data')

#call the xlabel function and assign the value 'Predicted values'
#call the ylabel function and assign the value 'Residuals'
#call the label function and for the loc parameter assign the value 'upper left'
#call the hlines function
##assign the y parameter the value '0'
##assign the xmin parameter the value '50'
##assign the color parameter the value 'black
##assign the lw parameter the value '2'
#call the xlim function and pass in the values [-10, 50]
#call the show function from the matplotlib library
plt.xlabel('Predicted values')
plt.ylabel('Residuals')
plt.legend(loc='upper left')
plt.hlines(y=0, xmin=-10, xmax=50, color='black', lw=2)
plt.xlim([-10, 50])
plt.show()
```
