```python
#create a dataframe with class imbalance
#0    23364
#1     6636

class_0 = df[df['Y'] == 0]
class_1 = df[df['Y'] == 1]
balanced_0 = class_0.iloc[:6636 , :]
balanced_1 = class_1.iloc[:6636, :]
balanced_df = balanced_0
balanced_df = balanced_df.append(balanced_1)
balanced_df['Y'].value_counts()
```

```python
#create our feature matrix and target vector
y = balanced_df.iloc[:, 21]
X = balanced_df.drop(['ID', 'Y', 'X12', 'X13', 'X14', 'X15', 'X16', 'X17', 'X18', 'X19', 'X20', 'X21', 'X22', 'X23'], axis=1)
```

```python
#perform feature extraction
from sklearn.decomposition import PCA

pca = PCA(n_components=None)
X_pca = pca.fit_transform(X)
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel('Number of Components')
plt.ylabel('Cumulative Explained Variance')
plt.show()
```

```python
#split the data into test and train sets
from sklearn.model_selection import train_test_split


X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=1)
```

```python
from sklearn.ensemble import RandomForestClassifier

feat_labels = X.columns[:]
forest = RandomForestClassifier(n_estimators=500,random_state=1)
forest.fit(X_train,y_train)
importances = forest.feature_importances_
indices = np.argsort(importances)[::-1]
for f in range(X_train.shape[1]):
    print('{}) {:<30} {:6f}'.format(f+1,feat_labels[indices[f]],importances[indices[f]]))

plt.title('Feature Importance')
plt.bar(range(X_train.shape[1]),importances[indices],align='center')
plt.xticks(range(X_train.shape[1]),feat_labels[indices],rotation=90)
plt.xlim([-1,X_train.shape[1]])
plt.tight_layout()
plt.show()
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import GridSearchCV

param_C = [0.001, 0.01, 0.1, 1.0, 10.0, 100.0, 1000.0]

lr_grid = [{'LR__penalty': ['l1', 'l2'],
            'LR__C': param_C,
            'LR__solver': ['liblinear']}]

pipe_lr = Pipeline([('scl', StandardScaler()),
                    ('LR', LogisticRegression(random_state=1))])

lr_grid_cv = GridSearchCV(estimator=pipe_lr, param_grid=lr_grid, scoring='accuracy', cv=10, n_jobs=-1)

lr_gs = lr_grid_cv.fit(X_train, y_train)
print('Logistic Regression Best Parameters: {}'.format(lr_gs.best_params_))
print('Logistic Regression Train Accuracy: {:.6f}'.format(lr_gs.best_score_))

lr_best_clf = lr_gs.best_estimator_
lr_best_clf.fit(X_train, y_train)

print('Logistic Regression Test Accuracy: {:.6f}'.format(lr_best_clf.score(X_test, y_test)))
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

y_hat = lr_gs.predict(X_test)
cm = confusion_matrix(y_test, y_hat, labels=lr_gs.classes_)
disp = ConfusionMatrixDisplay(confusion_matrix=cm)
disp.plot()
plt.show()
```

```python
from sklearn.model_selection import validation_curve

param_range = [0.001, 0.01, 0.1, 1.0, 10.0, 100.0]
train_scores, test_scores = validation_curve(estimator=pipe_lr, 
                                             X=X_train, y=y_train, 
                                             param_name='LR__C', 
                                             param_range=param_range,
                                             cv=10)
train_mean = np.mean(train_scores, axis=1)
train_std = np.std(train_scores, axis=1)
test_mean = np.mean(test_scores, axis=1)
test_std = np.std(test_scores, axis=1)
plt.plot(param_range, train_mean, 
         color='blue', marker='o', 
         markersize=5, 
         label='training accuracy')
plt.fill_between(param_range, train_mean + train_std,
                 train_mean - train_std, alpha=0.15,
                 color='blue')
plt.plot(param_range, test_mean, 
         color='green', linestyle='--', 
         marker='s', markersize=5, 
         label='validation accuracy')
plt.fill_between(param_range, 
                 test_mean + test_std,
                 test_mean - test_std, 
                 alpha=0.15, color='green')
plt.grid()
plt.xscale('log')
plt.legend(loc='lower right')
plt.xlabel('Parameter C')
plt.ylabel('Accuracy')
plt.ylim([0.8, 1.5])
plt.show() 
```

```python
from sklearn.tree import DecisionTreeClassifier

param_criterion = ['gini', 'entropy']
param_range = [1, 2, 3, 4, 5, 6]

DT_grid = [{'DT__criterion': param_criterion,
            'DT__max_depth': param_range}]

pipe_DT = Pipeline([('scl', StandardScaler()),
                    ('DT', DecisionTreeClassifier(random_state=1))])

DT_grid_cv = GridSearchCV(estimator=pipe_DT, param_grid=DT_grid, scoring='accuracy', cv=10, n_jobs=-1)

DT_gs = DT_grid_cv.fit(X_train, y_train)
print('Decision Tree Best Parameters: {}'.format(DT_gs.best_params_))
print('Decision Tree Train Accuracy: {:.6f}'.format(DT_gs.best_score_))

DT_best_clf = DT_gs.best_estimator_
DT_best_clf.fit(X_train, y_train)

print('Decision Tree Test Accuracy: {:.6f}'.format(DT_best_clf.score(X_test, y_test)))
```

```python
y_hat = DT_gs.predict(X_test)
cm = confusion_matrix(y_test, y_hat, labels=DT_gs.classes_)
disp = ConfusionMatrixDisplay(confusion_matrix=cm)
disp.plot()
plt.show()
```
